## 1.สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่:
  #### function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน โดย listแรกเป็น listของตัวเลข listที่สองเป็น listของ text
  #### ถ้าค่าใน listแรกเป็นเลขคี่ ให้ค่าในdictเป็นoutput มี key เป็นค่าจาก list ตัวที่1 และ value จาก listตัวที่2
  #### ถ้าค่าใน listแรกเป็นเลขคู่ ให้ค่าในdictเป็นoutput มี key เป็นค่าจาก list ตัวที่2 และ value จาก listตัวที่1
  #### output ของfunctionคือ dict ที่สร้างขึ้น
  ## เช่น
  #### output1=smartDict[1,2,4]),['apple','banana','orange'])
  #### จะได้ output1 เป็น1:'apple',banana':2,3:'orange'}


```python
def createDict(listnumber,listtext,):
    retdict = {}
    for i in range(len(listnumber)):
        if listnumber [i]%2 == 1 :
            retdict[listnumber[i]] = listtext[i]
        else:
             retdict[listtext[i]] = listnumber[i]
    return retdict
```

```python
createDict( [ 1 , 2 ,4],[ 'apple' , 'banana','orange' ] )
```

```python
def countChar(seq,character):
    charcount = 0
    for i in seq :
        if i == character :
            charcount = charcount + 1
    return charcount
```

```python
Aloc = 0
Tloc= 0
Gloc = 0
Cloc = 0
for i in range(DNAseq.shape[0]):
    Aloc += countChar(DNAseq.iloc[i][0],'A')
    Tloc += countChar(DNAseq.iloc[i][0],'T')
    Gloc += countChar(DNAseq.iloc[i][0],'G')
    Cloc += countChar(DNAseq.iloc[i][0],'C')
print(f'There are A= {Aloc} , T= {Tloc} , G= {Gloc} , C= {Cloc} ')

```

## ซ้ำกันมากที่สุด3ตัว

```python
def populate3char(seq):
    str_list = list()
    for i in range(len(seq)-2) :
        str_list.append(seq[i:i+3])
    return str_list
```

```python
populate3char('ATTA')
```

```python
list_all3char = list()
for row in range(DNAseq.shape[0]):
    list_all3char = list_all3char + populate3char(DNAseq.iloc[row][0])
```

```python
len(list_all3char)
```

## *กำจัด set ที่ซ้ำกัน

```python
unique_3char = list(set(list_all3char))
```

```python
len(unique_3char)
```

```python
dict_3char = {}
for sub_str in unique_3char :
    num = 0
    for s in list_all3char :
        if sub_str == s :
            num += 1
    dict_3char[sub_str] = num
```

```python
dict_3char
```

```python
dict_3char['TGGT']
```

```python
maxValue = -100
maxKey = None
for k in dict_3char.keys():
    if dict_3char[k] > maxValue:
        maxValue = dict_3char[k]
        maxKey = k
print(f'max 3-long-sequence is {maxKey} with frequency equals {maxValue}')
```
