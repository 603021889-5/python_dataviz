ลบข้อมูลของประเทศจีนและจีนแผ่นดินใหญ่ออกเนื่องจากค่าข้อมูลมันสูงกว่าค่าอื่นมากๆ

```python
covid = df[df['Country'] != 'China']
```

```python
covid = covid[covid['Country'] != 'Mainland China']
```

```python
covid = covid.reset_index(drop=True)
```

```python
covid
```

แก้ไข column date

```python
date = []
for i in range(covid.shape[0]):
    x = covid.iloc[i]['Date'].split(' ')
    x = x[0].replace('/2020', '')
    date.append(x)
```

```python
covid['Date'] = date
```

```python
covid
```

ลบ columns ที่ไม่ต้องการออก

```python
del covid['Sno']
del covid['Last Update']
```

```python
covid
```

```python
covid.to_csv('covid_19.csv')
```

### create graph

```python
x = covid.groupby(['Country'], as_index=False).sum()['Country']
y1 = covid.groupby(['Country'], as_index=False).sum()['Confirmed']
y2 = covid.groupby(['Country'], as_index=False).sum()['Deaths']
y3 = covid.groupby(['Country'], as_index=False).sum()['Recovered']
plt.figure(figsize=[10,5],dpi=80)
plt.gca(facecolor='oldlace')
plt.plot(x,y1,'blue',label = "Confirmed",marker='o')
plt.plot(x,y2,'red',label = "Deaths",marker='o')
plt.plot(x,y3,'green',label = "Recovered",marker='o')
plt.legend()
plt.xticks(rotation = 90)
plt.xlabel('ประเทศ',fontname='Tahoma')
plt.ylabel('จำนวน(คน)',fontname='Tahoma')
plt.title('Covid-19 2020')
```

```python
x = china.groupby(['Date'], as_index=False).sum()['Date']
y1 = china.groupby(['Date'], as_index=False).sum()['Confirmed']
y2 = china.groupby(['Date'], as_index=False).sum()['Deaths']
y3 = china.groupby(['Date'], as_index=False).sum()['Recovered']
plt.figure(figsize=[10,5],dpi=80)
plt.gca(facecolor='skyblue')
plt.plot(x,y1,'r',label = "Confirmed",marker='*')
plt.plot(x,y2,'b',label = "Deaths",marker='*')
plt.plot(x,y3,'y',label = "Recovered",marker='*')
plt.legend()
plt.xticks(rotation = 90)
plt.xlabel('วันที่',fontname='Tahoma')
plt.ylabel('จำนวน(คน)',fontname='Tahoma')
plt.title('Covid-19 in China')
```
