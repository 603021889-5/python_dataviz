```python
setosa_np=irisdata.iloc[:50,:-1].to_numpy()
setosa_np
```

## s แทน ขนาด marker(จุด)

## c แทน สีของจุด

## s = 10*setosa_np[:,2] คือ กำหนดให้ขนาดของจุดเท่ากับสิบเท่าของข้อมูลในหลักที่3

## marker แทน รูปแบบของ marker

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c='c')
```

```python
print(f'max={max(setosa_np[:,3])},min ={min(setosa_np[:,3])}')
```

## การกำหนดสีแต่ละช่วงค่าข้อมูล

## ถ้าค่าในหลักที่4 น้อยกว่า 0.15 ให้เป็นสีน้ำเงิน
## ถ้าค่าในหลักที่4 น้อยกว่า 0.30 ให้เป็นสีเขียว
## ถ้าค่าในหลักที่4 น้อยกว่า 0.45 ให้เป็นสีเหลือง
## ถ้าค่าในหลักที่4 มากกว่าหรือเท่ากับ 0.45 ให้เป็นสีแดง

```python
colors=[]
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.30:
        colors.append('g') 
    elif i < 0.45:
        colors.append('y') 
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c=colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c=setosa_np[:,3],cmap='YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c=np.array([1,0.5,0.8]))
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c=setosa_np[:,3],cmap='YlOrBr',marker='p')
```
