ข้อ1. สร้าง function ชื่อ createDict ซึ่งทำหน้าที่:

function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน และ ค่า True/False
ถ้า True จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
ถ้า False จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
output ของ function คือ dict ที่สร้างขึ้น
เช่น

output1 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )

จะได้ output1 เป็น { 'apple' : 1 , 'banana' : 2 }

output2 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )

จะได้ output2 เป็น { 1 : 'apple' , 2 : 'banana' }

```python
def createDict(listtext,listnumber,logicV):
    if logicV : #logicV == True
        retdict = {}
        for i in range(len(listtext)):
            retdict[listtext[i]] = listnumber[i]
    else : #logicV == False
        retdict = {}
        for i in range(len(listtext)):
            retdict[listnumber[i]] = listtext[i]
    return retdict
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )
```

ข้อ 3.2

```python
def countChar(seq,character):
    charcount = 0
    for i in seq :
        if i == character :
            charcount = charcount + 1
    return charcount
```

```python
A = 0
T = 0
G = 0
C = 0
for i in range(DNAseq.shape[0]):
    A += countChar(DNAseq.iloc[i][0],'A')
    T += countChar(DNAseq.iloc[i][0],'T')
    G += countChar(DNAseq.iloc[i][0],'G')
    C += countChar(DNAseq.iloc[i][0],'C')
print(f'There are {A} As, {T} Ts, {G} Gs, {C} Cs')
```

ข้อ 3.3

```python
def populate5char(seq):
    str_list = list()
    for i in range(len(seq)-4) :
        str_list.append(seq[i:i+5])
    return str_list
```

```python
populate5char('ATTAAA')
```

```python
list_all5char = list()
for row in range(DNAseq.shape[0]):
    list_all5char = list_all5char + populate5char(DNAseq.iloc[row][0])
```

```python
len(list_all5char)
```

```python
unique_5char = list(set(list_all5char))
```

***set กำจัดตัวซ้ำ

```python
len(unique_5char)
```

```python
dict_5char = {}
for sub_str in unique_5char :
    num = 0
    for s in list_all5char :
        if sub_str == s :
            num += 1
    dict_5char[sub_str] = num
```

```python
dict_5char
```

```python
dict_5char['TGGTG']
```

```python
maxValue = -100
maxKey = None
for k in dict_5char.keys():
    if dict_5char[k] > maxValue:
        maxValue = dict_5char[k]
        maxKey = k
print(f'max 5-long-sequence is {maxKey} with frequency equals {maxValue}')
```
